```python
# import required libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# use .get method to get data from url
url = 'https://thefishy.co.uk/leaguetable.php?table=1&season=15'
page = requests.get(url)

# check response code, should be 200 for successful connection
print(page.status_code)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
# find table based on class found by inspecting element in web browser
tb = soup.find('table', class_='table table-condensed table-striped table-bordered')
```

```python
# open context for csv file to store output
with open ('epl_stats.csv', 'w') as r:
    # loop through the rows in the table
    for row in tb.find_all('tr'):
        # find all th elements, which are column names and write to file, separate using comma
        # the comma writing will result in an extra row which will be removed when reading file
        # this can be improved later if time allows
        for cell in row.find_all('th'):
            r.write(cell.text)
            r.write(',')
        # find all td elements, which are cell values and write to file, separate using comma
        for cell in row.find_all('td'):
            r.write(cell.text)
            r.write(',')
        # add line break for each row
        r.write('\n')
```

```python
# read in the csv file we output above
epl_stats = pd.read_csv('data/epl_stats.csv')
# drop the extra column caused by the extra comma
#epl_stats.drop('Unnamed: 21', axis=1, inplace=True)
# set the index to equal the rank/position of the team
epl_stats.set_index('Pos', inplace=True)
```

```python
epl_stats.replace('Bournemouth', 'AFC Bournemouth', inplace=True)
# display the results
epl_stats['Position'] = epl_stats.index.values
epl_stats.sort_values('Team')
```

```python
teams = pd.read_csv('data/premier_league_teams.csv', index_col='teamId')
```

```python
now = epl_stats.Team
later = teams.name
compare = pd.DataFrame(now)
compare.sort_values('Team', inplace=True)
c2 = compare.reset_index()
l1 = pd.DataFrame(later).sort_values('name')
l2 = l1.reset_index()
c3 = c2.Team
l3 = l2.name
c2.index.values
l2.index.values
d = pd.concat([c3,l3], axis=1)
```

```python
epl_stats2 = epl_stats.merge(d, on='Team')
```

```python
epl_stats3 = epl_stats2[['name', 'P', 'HW', 'HD', 'HL', 'HF', 'HA', 'AW', 'AD', 'AL', 'AF', 'AA',
       'W', 'D', 'L', 'F', 'A', 'GD', 'Pts', 'PPG', 'Position']]
```

```python
epl_stats3.set_index('Position', inplace=True)
epl_stats3.head()
```

```python
epl_stats3.to_csv('data/epl_stats.csv', index=True, header=True)
```
